Tuổi thọ là một chỉ số quan trọng phản ánh tình trạng sức khỏe, điều kiện sống và chất lượng dịch vụ y tế của người dân trong một quốc gia. Các quốc gia có tuổi thọ cao thường đi kèm với tỷ lệ tử vong ở trẻ sơ sinh và trẻ nhỏ thấp, tỷ lệ tiêm chủng và khám chữa bệnh hiệu quả, điều kiện vệ sinh môi trường tốt cũng như thói quen dinh dưỡng và lối sống lành mạnh. Ngược lại, tuổi thọ thấp có liên quan đến đói nghèo, dịch bệnh, ô nhiễm môi trường và các vấn đề sức khỏe cộng đồng.

Việc phân tích các yếu tố ảnh hưởng đến tuổi thọ sẽ giúp các nhà hoạch định chính sách đưa ra các giải pháp nhằm cải thiện điều kiện sống và nâng cao chất lượng dịch vụ y tế cho người dân. Đồng thời, việc so sánh sự chênh lệch về tuổi thọ giữa các nước cũng cho phép đánh giá hiệu quả của các chính sách phát triển kinh tế, xã hội đang được áp dụng.

## Phát biểu vấn đề
Tuổi thọ là một chỉ số quan trọng không chỉ đánh giá sức khỏe cộng đồng mà còn ảnh hưởng đến sự phát triển kinh tế của mỗi quốc gia. Sự chênh lệch về tuổi thọ giữa các quốc gia không chỉ là một vấn đề y tế mà còn mang theo những thách thức lớn về mặt kinh tế và xã hội. Nghiên cứu này sử dụng các phương pháp phân tích số liệu để mở rộng kiến thức về những yếu tố ảnh hưởng đến tuổi thọ và thay đổi của chúng theo thời gian.

Các câu hỏi nghiên cứu chính:

* Yếu tố ảnh hưởng đến tuổi thọ: Chúng ta cần hiểu rõ những yếu tố nào có ảnh hưởng đến tuổi thọ của con người. Từ việc phân tích dữ liệu thu thập từ bộ dữ liệu "Life Expectancy" của Tổ chức Y tế Thế giới (WHO), chúng tôi mong muốn xác định những yếu tố quan trọng nhất trong việc định hình tuổi thọ trung bình.
* Thay đổi của các yếu tố theo thời gian: Chúng tôi quan tâm đến việc làm thế nào các yếu tố ảnh hưởng đến tuổi thọ thay đổi qua các năm. Việc theo dõi sự biến động này sẽ giúp chúng tôi hiểu rõ hơn về xu hướng và động lực đằng sau sự thay đổi của tuổi thọ trên toàn thế giới.
* Chính sách cải thiện tuổi thọ: Nghiên cứu này không chỉ là việc phân tích mà còn là cơ hội để đề xuất các chính sách có thể cải thiện tuổi thọ. Chúng tôi hy vọng rằng thông qua việc hiểu sâu hơn về những yếu tố chi phối, chúng tôi có thể đưa ra những đề xuất hữu ích để cải thiện sức khỏe cộng đồng toàn cầu.

Chúng tôi tập trung vào phân tích dữ liệu từ bộ dữ liệu để không chỉ đưa ra các nhận định mà còn đề xuất những biện pháp cụ thể để nâng cao chất lượng cuộc sống và tuổi thọ của cộng đồng toàn thế giới.

## Giới thiệu bộ dữ liệu
Bộ dữ liệu liên quan đến tuổi thọ, yếu tố sức khỏe của 193 quốc gia được thu thập từ cùng một trang web kho dữ liệu của WHO và dữ liệu kinh tế tương ứng được thu thập từ trang web của Liên hợp quốc, với sự hỗ trợ từ Deeksha Russell và Duan Wang.

Dữ liệu bao gồm các thuộc tính sau:

* Life Expectancy: Tuổi thọ trung bình

* Adult Mortality: Tỷ lệ tử vong người trưởng thành

* Polio: Số ca bệnh bại liệt

* Total Expenditure: Tổng chi tiêu cho y tế

* Diphtheria: Số ca bệnh bạch hầu

* HIV/AIDS: Số ca nhiễm HIV/AIDS

* GDP: GDP bình quân đầu người

* Population: Dân số

* Thinness 1-19 Years: Tỷ lệ suy dinh dưỡng trẻ em 1-19 tuổi

* Thinness 5-9 Years: Tỷ lệ suy dinh dưỡng trẻ em 5-9 tuổi

## Phân tích dữ liệu để trả lời các câu hỏi nghiên cứu

Q1. Tuổi thọ trung bình ở các quốc gia theo từng năm (Year) có xu hướng tăng hay giảm?

```python
# Tính tuổi thọ trung bình theo từng năm
avg_life_expectancy = df.groupby('Year')['Life Expectancy'].mean().reset_index()
# Vẽ biểu đồ đường
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Life Expectancy', data=avg_life_expectancy)
plt.title('Xu hướng Tuổi thọ trung bình qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tuổi thọ trung bình')
plt.show()
```

Nhận xét:

- Theo biểu đồ, có thể thấy tuổi thọ trung bình của các quốc gia có xu hướng tăng dần theo từng năm trong khoảng thời gian 2000-2015.

- Xu hướng tăng này phản ánh sự cải thiện về điều kiện sống, dinh dưỡng, y tế ở nhiều quốc gia trên thế giới, dẫn đến tuổi thọ trung bình tăng dần theo thời gian. Tuy nhiên, tốc độ tăng không đều giữa các nước.

Q2. Một quốc gia có giá trị tuổi thọ trung bình thấp hơn (<65) có nên tăng chi tiêu chăm sóc sức khỏe (Total Expenditure) để cải thiện tuổi thọ trung bình không?

```python
#Lọc ra các quốc gia có tuổi thọ trung bình nhỏ hơn 65
df_low_life_expectancy = df[df['Life Expectancy'] < 65]
#Vẽ đồ thị để hiển thị mối quan hệ giữa tổng chi tiêu chăm sóc sức khỏe và tuổi thọ trung bình
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Total Expenditure', y='Life Expectancy', data=df_low_life_expectancy)
plt.title('Mối quan hệ giữa Tổng chi tiêu Y tế và Tuổi thọ trung bình (Dưới 65 tuổi)')
plt.xlabel('Tổng chi tiêu Y tế')
plt.ylabel('Tuổi thọ trung bình')
plt.show()
# Kiểm tra tính phân phối của dữ liệu tuổi thọ trung bình dưới 65
stat, p_value = shapiro(df_low_life_expectancy['Life Expectancy'])
print(f'Statistic: {stat}, p-value: {p_value}')
# Nếu dữ liệu không tuân theo phân phối chuẩn, có thể sử dụng kiểm định Mann-Whitney U
result = mannwhitneyu(df_low_life_expectancy['Total Expenditure'], df_low_life_expectancy['Life Expectancy'])
print(f'Mann-Whitney U Statistic: {result.statistic}, p-value: {result.pvalue}')
```

Nhận xét:

- Statistic của kiểm định Shapiro là 0.9565, và p-value rất thấp (3.43e-15). Với mức ý nghĩa thống kê thấp, chúng ta có đủ chứng cứ để bác bỏ giả định về tính phân phối chuẩn của dữ liệu tuổi thọ trung bình dưới 65.

- Kiểm Định Mann-Whitney U: Statistic của kiểm định Mann-Whitney U là 0.0, và p-value rất gần 0 (2.01e-279). Với mức ý nghĩa thống kê rất thấp, chúng ta có đủ chứng cứ để bác bỏ giả định không có sự khác biệt giữa chi tiêu chăm sóc sức khỏe và tuổi thọ trung bình dưới 65.

- Nhận Xét Chung: Có sự tương quan đáng kể giữa chi tiêu chăm sóc sức khỏe và tuổi thọ trung bình ở nhóm quốc gia có giá trị tuổi thọ dưới 65. Các quốc gia có chi tiêu chăm sóc sức khỏe cao có xu hướng có tuổi thọ trung bình cao hơn so với nhóm có chi tiêu thấp.

Q3. Tỷ lệ tử vong ở trẻ sơ sinh (Infant Deaths) và người lớn (Adult Mortality) ảnh hưởng đến tuổi thọ như thế nào?

```python
plt.figure(figsize=(12, 5))
# Biểu đồ phân tán để quan sát mối tương quan giữa tuổi thọ trung bình và tỷ lệ tử vong ở trẻ sơ sinh
plt.subplot(1, 2, 1)
sns.scatterplot(x='Infant Deaths', y='Life Expectancy', data=df)
plt.xlabel('Tử vong ở trẻ sơ sinh')
plt.ylabel('Tuổi thọ trung bình')
plt.title('Tử vong ở trẻ sơ sinh vs Tuổi thọ trung bình')
# Biểu đồ phân tán để quan sát mối tương quan giữa tuổi thọ trung bình và tỷ lệ tử vong ở người lớn
plt.subplot(1, 2, 2)
sns.scatterplot(x='Adult Mortality', y='Life Expectancy', data=df)
plt.xlabel('Tử vong ở người lớn')
plt.ylabel('Tuổi thọ trung bình')
plt.title('Tử vong ở người lớn vs Tuổi thọ trung bình')
plt.tight_layout()
plt.show()
```

- Kết quả phân tích mô hình hồi quy cho thấy các yếu tố kinh tế - xã hội như thu nhập bình quân và chất lượng giáo dục có tương tác với tuổi thọ trung bình của một quốc gia. Trong đó, thu nhập bình quân đóng vai trò quan trọng nhất trong việc quyết định tuổi thọ. Những quốc gia có thu nhập cao và chất lượng giáo dục tốt thường có tuổi thọ trung bình dân số cao hơn. Q18. Có những biện pháp nào có thể được đề xuất để cải thiện tuổi thọ dựa trên các yếu tố quan trọng đã xác định từ dữ liệu?

```python
# Chuyển đổi dữ liệu thành kiểu numeric, xử lý các giá trị không hợp lệ thành NaN
df = df.apply(pd.to_numeric, errors='coerce')
# Tính ma trận tương quan và chọn cột 'Life Expectancy'
c = df.corr()['Life Expectancy']
# Lọc các biến có tương quan mạnh với Tuổi thọ (|correlation| > 0.5)
high_corr_vars = c[abs(c) > 0.5].sort_values(ascending=False)
# Hiển thị các biến có tương quan mạnh
print(high_corr_vars)
# Vẽ biểu đồ các biến có tương quan mạnh nhất với Tuổi thọ
plt.figure(figsize=(12, 6))
sns.barplot(x=high_corr_vars.index, y=high_corr_vars)
plt.xticks(rotation=90)
plt.title('Biến có tương quan mạnh nhất với Tuổi thọ')
plt.tight_layout()
plt.show()
```

Nhận xét:

- Dựa trên kết quả tương quan từ dữ liệu, có thể đề xuất những biện pháp nhất quán để cải thiện tuổi thọ. Nâng cao mức độ giáo dục thông qua đầu tư vào hệ thống giáo dục, cải thiện thu nhập và điều kiện sống, thúc đẩy chăm sóc sức khỏe và dinh dưỡng, kiểm soát HIV/AIDS, cũng như giảm tử vong người trưởng thành, là những hướng đi quan trọng. Những biện pháp này có thể tăng cường sức khỏe và chất lượng cuộc sống, góp phần vào việc nâng cao tuổi thọ của cộng đồng.
